```python
import pandas as pd
data = pd.DataFrame({'data':[1,2,3,3,3,6],'group' :[2,3,4,5,6,7]})
data
```

```python
data.sort_values(by = 'data',ascending=False)
```

#### 去除重复的数据，默认是完全相同，可以根据subset参数

```python
data.drop_duplicates(subset='data')
```

#### apply()，需要自己指定一个函数，apply会对DataFrame统一执行此函数，需要指定维度

```python
def food_map(series):
    if series['food'] == 'A1':
        return 'A'
    elif series['food'] == 'A2':
        return 'A'
    elif series['food'] == 'B1':
        return 'B'
    elif series['food'] == 'B2':
        return 'B'
    elif series['food'] == 'A3':
        return 'A'
    elif series['food'] == 'C1':
        return 'C'
    elif series['food'] == 'C2':
        return 'C'
```

```python
data = pd.DataFrame({'data' : [1,2,3,4,5,6,7], 'food' : ['A1','A2','A3','B1','B2','C1','C2']})
data
```

```python
data['food_map'] = data.apply(food_map, axis = 1)
data
```

#### map()，需要自己指定字典，对DataFrame中特定列按照字典返回映射值

```python
food2Upper = {
    'A1' : 'A',
    'A2' : 'A',
    'A3' : 'A',
    'B1' : 'B',
    'B2' : 'B',
    'C1' : 'C',
    'C2' : 'C',
}
```

```python
data['upper'] = data['food'].map(food2Upper)
data
```
